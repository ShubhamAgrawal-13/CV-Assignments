## 1. Put 4 images of your choice in the src folder.

```python
files = ['itachi_uchiha.png', 'levi_AOT.jpg', 'rengoku.png', 'Zenitsu_1.jpg']
# !ls
```

## 2. Load the images using OpenCV.

```python
images = []
for filename in files:
    images.append( cv2.imread(filename) )
```

## 3. Display the images using matplotlib (make a 2x2 grid using subplots). Do the images seem unusual? Fix the issue. 

```python
f, axarr = plt.subplots(2,2)
axarr[0,0].imshow(images[0])
axarr[0,1].imshow(images[1])
axarr[1,0].imshow(images[2])
axarr[1,1].imshow(images[3])
```

        The images look bluelish.
        This is because, by default, OpenCV reads image in the sequence Blue Green Red (BGR). 
        Thus, to view the actual image we need to convert the rendering to Red Green Blue (RGB).

```python
for i in range(len(images)):
    images[i] = cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB)
```

```python
f, axarr = plt.subplots(2,2)
axarr[0,0].imshow(images[0])
axarr[0,1].imshow(images[1])
axarr[1,0].imshow(images[2])
axarr[1,1].imshow(images[3])
```

```python
image = images[2]
print(image)
print(image.shape)
```

## 4. Convert any colour image (of your choice) to grayscale without using a function and save that image in the images folder.

```python
gray_image = []

for i in range(image.shape[0]) :
    row=[]
    for j in range(image.shape[1]):
        row.append(image[i,j,0]*0.07 + image[i,j,1]*0.72 + image[i,j,2] *0.21)
    gray_image.append(row)

gray_image = np.array(gray_image)
```

```python
plt.imshow(gray_image, cmap='gray')
```

```python
cv2.imwrite("../images/grayed_image.png" , gray_image.reshape(image.shape[0], image.shape[1]))
```

## 2. WebCam

```python
import cv2

camera = cv2.VideoCapture(0)

return_value, image = camera.read()
plt.imshow(image)
cv2.imwrite('../images/opencv_cam_shot'+'.png', image)

del(camera)
```
